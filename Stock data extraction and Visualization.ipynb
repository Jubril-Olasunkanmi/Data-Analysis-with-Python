```python
import yfinance as yf
import pandas as pd
import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
samsung = yf.Ticker("SSNLF")
Apple = yf.Ticker("AAPL")
Panasonic = yf.Ticker("PCRFY")
Sony = yf.Ticker("SONY")

Products = [samsung, Apple, Panasonic, Sony]

for i in Products:
    Product_info = i.info
    print(Product_info)
```

```python
#Get Country names for each brand
for i in Products:
    Product_info = i.info
    print(Product_info['country'])
```

# Extracting Share Price
A share is the single smallest part of a company's stock that you can buy, the prices of these shares fluctuate over time. Using the history() method we can get the share price of the stock over a certain period of time. Using the period parameter we can set how far back from the present to get data. The options for period are 1 day (1d), 5d, 1 month (1mo) , 3mo, 6mo, 1 year (1y), 2y, 5y, 10y, ytd, and max.

```python
samsung_share_price_data = Products[0].history(period="max")
Apple_share_price_data = Products[1].history(period="max")
Panasonic_share_price_data = Products[2].history(period="max")
Sony_share_price_data = Products[3].history(period="max")

share_price_data = [samsung_share_price_data, Apple_share_price_data, Panasonic_share_price_data, Sony_share_price_data ]

for j in share_price_data:
    display(j.head())
```

```python
share_price_data = [samsung_share_price_data, Apple_share_price_data, Panasonic_share_price_data, Sony_share_price_data ]
for l in share_price_data:
    l.reset_index(inplace=True)
```

# Plotting the Open price against the Date

```python
share_price_data = [samsung_share_price_data, Apple_share_price_data, Panasonic_share_price_data, Sony_share_price_data ]

for k in share_price_data:
    k.plot(x="Date", y="Open", figsize = (13,7))
```

```python
for k in share_price_data:
    k.plot(x="Date", y="Dividends", figsize = (14, 7))
```
